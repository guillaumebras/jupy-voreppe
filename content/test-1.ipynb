```python
# Read the CSV file
# lecture de l'export CSV pour Voreppe (data.gouv) 2000 => 2022
vor_data = pd.read_csv("vor.csv", sep=";")

# View the first 5 rows
vor_data.head()

# Graph des emprunts sur le budget d'invest - on voit le 6ME de ~2021
alt.Chart(vor_data).mark_bar().encode(
    x='an',
    #y1='caf',
    y='emp'
)
```

```python


alt.Chart(vor_data).transform_fold(
  ["prod", "charge", "perso", "res1"],
  as_=["field", "value"]
).mark_bar().encode(
  x="field:N",
  y="sum(value):Q",
  column="an:N"
)
```

```python
# ebf = excÃ©dent brut
# caf = capacitÃ© d'autofinancement
# emp = emprunt (invest)
# depinv = dÃ©penses d'investissement

alt.Chart(vor_data).transform_fold(
  ["ebf", "caf", "depinv", "emp"],
  as_=["field", "value"]
).mark_bar().encode(
  x="field:N",
  y="sum(value):Q",
  column="an:N"
)
```
